ANS : Pandas is a powerful library for data manipulation in Python. Here are some common functions you can use to manipulate data in a Pandas DataFrame:

1.unique(): This function returns an array of unique values in the DataFrame.

2.nunique(): This function returns the number of unique values in the DataFrame.

3.describe(): This function returns a summary of statistics for each column in the DataFrame.

4.sort_values(): This function sorts the DataFrame by one or more columns.

5.value_counts(): This function returns a count of unique values in the DataFrame.

6.isnull(): This function returns a boolean array indicating which values are null (NaN).

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

ANS : According to Pandas in Python deals with three data structures namely Series DataFrame Panel. Of these, Series is a one-dimensional array with homogeneous data. All the elements of series should be of the same data type. It is size immutable & the size cannot be changed. DataFrames are two-dimensional arrays with heterogeneous data, usually represented in tabular format. The data is represented in rows and columns. Each column represents an attribute and each row represents a person. It is size mutable & the size can be changed. Panel is a three-dimensional data structure with heterogeneous data. It is size mutable & the size can be changed.
